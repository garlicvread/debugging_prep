Find the median.

- The number of elements is always odd number.

```python
def getMedian(data) :
    data = sorted(data)
    median_value = data[len(data)/2]
    return median_value

data = [int(v) for v in input().split()]

print(getMedian(data))
```

```python
# Solution

def getMedian(data) :
    data = sorted(data)
    median_value = data[int(len(data)/2)]
    return median_value

data = [int(v) for v in input().split()]

print(getMedian(data))
```
