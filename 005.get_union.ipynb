Get union of two given sets.
- Input example:<br>
1 2 3 4 5<br>
3 4 5 6 7

- Output example:<br>
1 2 3 4 5 6 7

```python
# Problem

oneLine = [int(v) for v in input().split()]
A = set(oneLine)

oneLine = [int(v) for v in input().split()]
B = set(oneLine)

Total = A + B

print(Total)
```

```python
# Solution

oneLine = [int(v) for v in input().split()]
A = set(oneLine)

oneLine = [int(v) for v in input().split()]
B = set(oneLine)


temp = str(A.union(B))


characters_to_remove = "{,}"
Total = ''.join( x for x in temp if x not in characters_to_remove)

print(Total)
```
