Addition and subtraction with exact integers.

```python
# Problem

def add(n1, n2):
    return n1 + n2
    
def subtract(n1, n2):
    return n1 - n2
    
if add(0.7, 0.2) == 0.9:
    print("add function now works fine!")
else:
    print("add function has been fucked!")
    
if subtract(1.2, 0.1) == 1.1:
    print("subtract function now works fine!")
else:
    print("subtractfunction has been fucked!!")
```

```python
# Solution

import math
import sys

def add(n1, n2):
    return n1 + n2
    
def subtract(n1, n2):
    return n1 - n2
    
    
x = add(0.7, 0.2)
if math.fabs(x - 0.9) <= sys.float_info.epsilon:
    print("add function now works fine!")
else:
    print("add function has been fucked!")
    
    
y = subtract(1.2, 0.1)
if math.fabs(y - 1.1) <= sys.float_info.epsilon:
    print("subtract function now works fine!")
else:
    print("subtractfunction has been fucked!!")
    
```
