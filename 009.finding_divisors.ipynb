```python
# Solution

def solve(n) :
    result = []
    
    for i in range(1, n+1) :
        mod = n % i
        if mod == 0 :
            result.append(i)
            
    return result


def main():
    n = int(input())
    
    print(*solve(n))


if __name__ == "__main__":
    main()
```
