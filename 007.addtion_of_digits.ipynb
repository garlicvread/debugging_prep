Add digits until the result becomes one digit value.

```python
# Problem


def solve(n) :
    result = 0
    
    while n >= 0 :
        result += n%10
        n //= 10
    
    if result >= 10 :
        solve(result)
    
    return result
    
def main() :
    n = int(input())
    
    print(solve(n))
    
if __name__ == "__main__" :
    main()
```

```python
# Solution


def solve(n) :
    result = n // 10 + n % 10

    while result >= 10:
        result = result // 10 + result % 10
        
    return(result)
    
    
def main() :
    n = int(input())
    
    print(solve(n))
    
    
if __name__ == "__main__" :
    main()
```
