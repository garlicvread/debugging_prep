Get the highest 3 scorers from the test.

- input:<br>
10 20 70 30 40 50 80 100<br>

- Output: <br>
8 7 3

```python
# Problem


def process(data) :
    data = data.sort()
    
    return [data[0][0], data[1][0], data[2][0]]
    
def main() :
    data = []
    
    scores = [int(v) for v in input().split()]
    
    for i in range(len(scores)) :
        data.append((i, scores[i]))
    
    print(*process(data))
    

if __name__ == "__main__" :
    main()
```

```python
# Solution

def process(data):
    answer = []

    for i in range(3):
        answer.append(data[i][0])

    return answer


def main():
    scores = [int(v) for v in input().split()]
    data = []

    for i in range(len(scores)):
        data.append([i+1, scores[i]])

    data.sort(key=lambda x: x[1], reverse=True)

    print(*process(data))
    
    
if __name__ == "__main__":
    main()
```
